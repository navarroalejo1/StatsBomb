# 3.Extraer cada partido de la competencia


```python

# Definir la ruta base donde se guardarán los archivos CSV
ruta_base = "C:/Users/anavarro/OneDrive - INDEPORTES ANTIOQUIA/AI/2024/METODOLOGO/PYTHON/MODULO 5/M5_Actividad_colaborativa/StatsBomb_America/"

# Verificar si la carpeta existe, si no, crearla
os.makedirs(ruta_base, exist_ok=True)

# Iterar a través de cada partido y obtener sus datos de eventos
for index, match in C_Am.iterrows():
    match_id = match['match_id']
    match_date = pd.to_datetime(match['match_date']).strftime('%Y-%m-%d')  # Formatear la fecha
    home_team = match['home_team']
    away_team = match['away_team']
    
    # Obtener los eventos del partido específico
    events = sb.events(match_id=match_id)
    
    # Convertir los datos de eventos a un DataFrame
    df_events = pd.DataFrame(events)
    
    # Crear un nombre de archivo basado en la fecha del partido y los nombres de los equipos
    nombre_archivo = f"{match_date}_{home_team}_vs_{away_team}.csv"
    ruta_archivo = os.path.join(ruta_base, nombre_archivo)
    
    # Guardar los eventos en un archivo CSV separado
    df_events.to_csv(ruta_archivo, index=False, encoding='utf-8')
    
    print(f"Archivo CSV para el partido {match_id} guardado en: {ruta_archivo}")

print("Todos los archivos CSV han sido generados.")
```

Utiliza .apply() para crear la columna foul_committed_card basada en los datos de bad_behaviour_card para cada jugador.

```python

# 1. Definir el ID del partido
match_id = 3943077

# 2. Obtener las alineaciones del partido específico
lineups = sb.lineups(match_id)

# 3. Obtener detalles del partido para el nombre del archivo
C_Am = sb.matches(competition_id=223, season_id=282)
match_info = C_Am[C_Am['match_id'] == match_id].iloc[0]
match_date = pd.to_datetime(match_info['match_date']).strftime('%Y-%m-%d')
home_team = match_info['home_team']
away_team = match_info['away_team']

# 4. Crear el nombre del archivo Excel usando match_id y match_date
nombre_archivo = f"{match_id}_{match_date}.xlsx"
ruta_base = "C:/Users/anavarro/OneDrive - INDEPORTES ANTIOQUIA/AI/2024/METODOLOGO/PYTHON/MODULO 5/M5_Actividad_colaborativa/StatsBomb_titulares"
ruta_archivo = os.path.join(ruta_base, nombre_archivo)

# Crear la carpeta de destino si no existe
os.makedirs(ruta_base, exist_ok=True)

# 5. Obtener los eventos del partido
events = sb.events(match_id=match_id)

# Verificar las columnas disponibles
print("Columnas disponibles en el DataFrame 'events':")
print(events.columns.tolist())

# Filtrar eventos que tienen tarjetas en 'bad_behaviour_card'
bad_behaviour_events = events[events['bad_behaviour_card'].notna()]

# 6. Preparar el archivo Excel con las alineaciones
with pd.ExcelWriter(ruta_archivo) as writer:
    for team_name, team_lineup in lineups.items():
        # Filtrar las columnas necesarias y extraer la posición inicial
        team_lineup_filtered = team_lineup[['player_id', 'country', 'player_name', 'jersey_number', 'positions']].copy()
        
        # Extraer la posición inicial de cada jugador
        team_lineup_filtered['initial_position'] = team_lineup_filtered['positions'].apply(
            lambda positions: positions[0]['position'] if isinstance(positions, list) and len(positions) > 0 else None
        )
        team_lineup_filtered.drop(columns=['positions'], inplace=True)  # Eliminar la columna original 'positions'

        # Crear columna 'foul_committed_card' con el conteo de "Yellow Card" o "Red Card" y su timestamp
        team_lineup_filtered['foul_committed_card'] = team_lineup_filtered['player_id'].apply(
            lambda x: ', '.join(
                [f"{card}- {pd.to_datetime(time, unit='s').strftime('%M:%S')}" 
                 for card, time in zip(bad_behaviour_events[bad_behaviour_events['player'] == x]['bad_behaviour_card'], 
                                       bad_behaviour_events[bad_behaviour_events['player'] == x]['timestamp'])]
            )
        )

        # Guardar la alineación del equipo en una hoja del archivo Excel
        team_lineup_filtered.to_excel(writer, sheet_name=team_name, index=False)

# Confirmación del guardado del archivo
print(f"\nArchivo Excel guardado en: {ruta_archivo}")
```

```python


# 1. Obtener todos los partidos de la competencia y la temporada específicas
C_Am = sb.matches(competition_id=223, season_id=282)

# 2. Iterar sobre cada partido en la competencia
for index, match_info in C_Am.iterrows():
    match_id = match_info['match_id']
    match_date = pd.to_datetime(match_info['match_date']).strftime('%Y-%m-%d')
    home_team = match_info['home_team']
    away_team = match_info['away_team']

    # Crear el nombre del archivo Excel usando match_id y match_date
    nombre_archivo = f"{match_id}_{match_date}.xlsx"
    ruta_base = "C:/Users/anavarro/OneDrive - INDEPORTES ANTIOQUIA/AI/2024/METODOLOGO/PYTHON/MODULO 5/M5_Actividad_colaborativa/StatsBomb_titulares"
    ruta_archivo = os.path.join(ruta_base, nombre_archivo)

    # Crear la carpeta de destino si no existe
    os.makedirs(ruta_base, exist_ok=True)

    # 3. Obtener las alineaciones del partido específico
    lineups = sb.lineups(match_id)

    # 4. Obtener los eventos del partido
    events = sb.events(match_id=match_id)

    # Verificar si 'bad_behaviour_card' existe en las columnas de 'events'
    if 'bad_behaviour_card' in events.columns:
        # Filtrar eventos que tienen tarjetas en 'bad_behaviour_card'
        bad_behaviour_events = events[events['bad_behaviour_card'].notna()]
    else:
        bad_behaviour_events = pd.DataFrame()  # Crear un DataFrame vacío si la columna no existe

    # 5. Preparar el archivo Excel con las alineaciones
    with pd.ExcelWriter(ruta_archivo) as writer:
        for team_name, team_lineup in lineups.items():
            # Filtrar las columnas necesarias y extraer la posición inicial
            team_lineup_filtered = team_lineup[['player_id', 'country', 'player_name', 'jersey_number', 'positions']].copy()
            
            # Extraer la posición inicial de cada jugador
            team_lineup_filtered['initial_position'] = team_lineup_filtered['positions'].apply(
                lambda positions: positions[0]['position'] if isinstance(positions, list) and len(positions) > 0 else None
            )
            team_lineup_filtered.drop(columns=['positions'], inplace=True)  # Eliminar la columna original 'positions'

            # Crear columna 'foul_committed_card' solo si hay eventos de tarjetas
            if not bad_behaviour_events.empty:
                team_lineup_filtered['foul_committed_card'] = team_lineup_filtered['player_id'].apply(
                    lambda x: ', '.join(
                        [f"{card}- {pd.to_datetime(time, unit='s').strftime('%M:%S')}" 
                         for card, time in zip(bad_behaviour_events[bad_behaviour_events['player'] == x]['bad_behaviour_card'], 
                                               bad_behaviour_events[bad_behaviour_events['player'] == x]['timestamp'])]
                    )
                )
            else:
                team_lineup_filtered['foul_committed_card'] = ""  # Dejar en blanco si no hay tarjetas

            # Guardar la alineación del equipo en una hoja del archivo Excel
            team_lineup_filtered.to_excel(writer, sheet_name=team_name, index=False)

    # Confirmación del guardado del archivo
    print(f"\nArchivo Excel guardado en: {ruta_archivo}")
```

 Se utiliza pd.ExcelWriter(ruta_archivo) para abrir un contexto que permite escribir en un archivo Excel.
 Se inicia un bucle for para iterar a través de cada equipo único en events. Para cada equipo, se filtran sus eventos específicos.

7. Filtrar eventos para el equipo específico:

 Se filtran los eventos en events para obtener solo aquellos que pertenecen al equipo actual (team_name).
 Si no hay eventos para el equipo, se imprime un mensaje y se pasa al siguiente equipo con continue.

8. Seleccionar las columnas relevantes:

 Se crea un nuevo DataFrame selected_columns que solo contiene las columnas necesarias para el análisis: 'team', 'player', 'pass_cross', y 'type'.

 Se cuentan los eventos específicos (Ball Recovery, Shot, Foul Committed) y los pases cruzados (pass_cross con valor VERDADERO) para cada jugador.

10. Crear el DataFrame individual_stats combinando los conteos:

 - Se combinan los DataFrames resultantes de los conteos (ball_recovery_count, shot_count, foul_committed_count, pass_cross_count) en un solo DataFrame llamado individual_stats.

11. Añadir la columna 'team':

 Se añade una columna 'team' a individual_stats para indicar a qué equipo pertenece cada jugador.

12. Rellenar valores nulos con ceros:

 Se utilizan ceros para reemplazar cualquier valor nulo en individual_stats para asegurar que todos los jugadores tengan valores consistentes.

13. Guardar las estadísticas en una hoja del archivo Excel:

 Se guarda el DataFrame individual_stats en una hoja del archivo Excel con el nombre del equipo como nombre de la hoja. Se imprime un mensaje para confirmar el guardado.

 Se imprime un mensaje final para confirmar que el archivo Excel ha sido guardado correctamente en la ubicación especificada.

```python

# Paso 1: Identificar el match_id y cargar los eventos del partido
match_id = 3943077  # Cambia el match_id al que necesites
events = sb.events(match_id=match_id)

# Verificar las columnas disponibles en el DataFrame 'events'
print("Columnas disponibles en el DataFrame 'events':")
print(events.columns.tolist())

# Paso 2: Crear un DataFrame para cada equipo y calcular estadísticas

# Definir la ruta base y crear la carpeta si no existe
ruta_base = "C:/Users/anavarro/OneDrive - INDEPORTES ANTIOQUIA/AI/2024/METODOLOGO/PYTHON/MODULO 5/M5_Actividad_colaborativa/StatsBomb_players"
os.makedirs(ruta_base, exist_ok=True)  # Crear la carpeta si no existe

# Crear el nombre del archivo Excel usando match_id
nombre_archivo = f"{match_id}_individual_stats.xlsx"
ruta_archivo = os.path.join(ruta_base, nombre_archivo)

# Preparar el archivo Excel para guardar las estadísticas
with pd.ExcelWriter(ruta_archivo) as writer:
    for team_name in events['team'].unique():
        print(f"\nProcesando datos para el equipo: {team_name}")

        # Filtrar eventos para el equipo específico
        team_events = events[events['team'] == team_name]

        # Verificar si el DataFrame 'team_events' no está vacío
        if team_events.empty:
            print(f"No hay eventos para el equipo {team_name} en este partido.")
            continue

        # Seleccionar las columnas relevantes
        selected_columns = team_events[['team', 'player', 'pass_cross', 'type']]

        # Contar eventos específicos por jugador
        ball_recovery_count = selected_columns[selected_columns['type'] == 'Ball Recovery'].groupby('player').size().reset_index(name='ball_recovery_count')
        shot_count = selected_columns[selected_columns['type'] == 'Shot'].groupby('player').size().reset_index(name='shot_count')
        foul_committed_count = selected_columns[selected_columns['type'] == 'Foul Committed'].groupby('player').size().reset_index(name='foul_committed_count')

        # Contar "VERDADERO" en "pass_cross" por jugador
        pass_cross_count = selected_columns[selected_columns['pass_cross'] == True].groupby('player').size().reset_index(name='pass_cross_true_count')

        # Crear el DataFrame 'individual_stats' combinando los conteos
        individual_stats = pd.merge(ball_recovery_count, shot_count, on='player', how='outer')
        individual_stats = pd.merge(individual_stats, foul_committed_count, on='player', how='outer')
        individual_stats = pd.merge(individual_stats, pass_cross_count, on='player', how='outer')

        # Añadir columna de 'team' al DataFrame resultante
        individual_stats['team'] = team_name

        # Rellenar valores nulos con ceros en caso de que algunos jugadores no tengan eventos específicos
        individual_stats.fillna(0, inplace=True)

        # Guardar las estadísticas combinadas en una hoja del archivo Excel
        individual_stats.to_excel(writer, sheet_name=team_name, index=False)

        print(f"Estadísticas guardadas para {team_name}.")

# Confirmación del guardado del archivo
print(f"\nArchivo Excel guardado en: {ruta_archivo}")
```

 Utilizamos la función sb.matches() de la biblioteca statsbombpy para obtener todos los partidos de la competencia y temporada especificadas. El resultado se guarda en el DataFrame C_Am.

 Definimos la variable ruta_base_pdf como la ruta donde se guardará el PDF generado. Usamos os.makedirs() para crear la carpeta si no existe.

4. eleccionar y extraer información del partido:

Seleccionamos un partido específico para analizar (match_id = 3943077).
match_info extrae la fila correspondiente a este match_id de C_Am.
Obtenemos información detallada del partido, como la fecha (match_date), los equipos locales y visitantes (home_team, away_team), los puntajes (home_score, away_score), y la etapa de la competencia (competition_stage).

5. Obtener alineaciones y eventos del partido:

sb.lineups(match_id) obtiene las alineaciones de ambos equipos para el partido especificado.
sb.events(match_id=match_id) obtiene todos los eventos del partido (como pases, tiros, faltas, etc.).

6. Filtrar eventos de tarjetas de comportamiento incorrecto:

Verificamos si la columna 'bad_behaviour_card' existe en el DataFrame events.
Si existe, filtramos los eventos que tienen tarjetas en esta columna.
Si no existe, creamos un DataFrame vacío para evitar errores posteriores.

7. Crear DataFrame vacío: 

Inicializamos all_data_df como un DataFrame vacío que se usará para almacenar las alineaciones y estadísticas individuales procesadas para cada equipo.

8. Procesar alineaciones de cada equipo:

Iteramos sobre cada equipo (team_name, team_lineup) en lineups.
Añadimos el nombre del equipo a cada fila (team).
Seleccionamos las columnas relevantes (player_id, country, player_name, jersey_number, positions, team).
Extraemos la posición inicial de cada jugador usando una función lambda que verifica si hay posiciones y toma la primera posición.
Eliminamos la columna original 'positions' después de extraer la posición inicial.
Si hay eventos de tarjetas, creamos la columna 'foul_committed_card' que concatena la tarjeta y el tiempo de evento.

9. Filtrar eventos y contar estadísticas:

Filtramos los eventos solo para el equipo específico (team_events).
Contamos eventos específicos por jugador:
Ball Recovery (recuperación de balón).
Shot (tiros).
Foul Committed (faltas cometidas).
Contamos cuántos pases cruzados (pass_cross) fueron verdaderos por jugador.

10. Unir alineaciones y estadísticas individuales:

Combinamos las estadísticas calculadas (individual_stats) con las alineaciones del equipo (team_lineup_filtered) usando un merge.
Eliminamos la columna duplicada 'player' y mantenemos solo 'player_name'.
Rellenamos valores nulos con ceros para evitar problemas con datos faltantes.
Eliminamos la columna 'country' si es igual a la información de la columna 'team'.

11. Ordenar y preparar datos para el PDF:

Convertimos 'initial_position' a cadena y reemplazamos valores nan con cadenas vacías.
Ordenamos por 'initial_position' (primero los que tienen posición, luego los vacíos) y 'player_name'.
Añadimos los datos procesados al DataFrame general all_data_df.

12. Crear y configurar el PDF:
Creamos un objeto PDF en modo horizontal (orientation='L') con tamaño A4.
Configuramos el salto de página automático con un margen de 15 mm.

13. Generar hoja en el PDF para cada equipo:

Iteramos por cada equipo en all_data_df y generamos una página nueva.
Añadimos el nombre del equipo como encabezado en negrita y tamaño 14.
Añadimos un encabezado de partido con detalles específicos (tamaño 8).
Ajustamos dinámicamente el ancho de las columnas según el número de columnas para evitar superposiciones.

14. Guardar el archivo PDF:

Definimos el nombre del archivo PDF usando el match_id.
Combinamos la ruta base y el nombre del archivo para obtener la ruta completa.
Utilizamos pdf.output para guardar el PDF en la ruta especificada.

```python

# 1. Definir el ID de la competencia y la temporada
competition_id = 223
season_id = 282

# 2. Obtener todos los partidos de la competencia y temporada específica
C_Am = sb.matches(competition_id=competition_id, season_id=season_id)

# 3.Definir la ruta base para guardar el PDF
ruta_base_pdf = "C:/Users/anavarro/OneDrive - INDEPORTES ANTIOQUIA/AI/2024/METODOLOGO/PYTHON/MODULO 5/M5_Actividad_colaborativa/StatsBomb_America_equipo"
os.makedirs(ruta_base_pdf, exist_ok=True)  # Crear la carpeta si no existe

# 4. Seleccionar un partido específico para analizar (match_id = 3943077)
match_id = 3943077
match_info = C_Am[C_Am['match_id'] == match_id].iloc[0]
match_date = pd.to_datetime(match_info['match_date']).strftime('%Y-%m-%d')
home_team = match_info['home_team']
away_team = match_info['away_team']
home_score = match_info['home_score']
away_score = match_info['away_score']
competition_stage = match_info['competition_stage']

# 5. Obtener las alineaciones y eventos del partido
lineups = sb.lineups(match_id)
events = sb.events(match_id=match_id)

# 6.Verificar si 'bad_behaviour_card' existe en las columnas de 'events'. 
# Filtrar Tarjetas de Comportamiento Incorrecto
if 'bad_behaviour_card' in events.columns:
    # Filtrar eventos que tienen tarjetas en 'bad_behaviour_card'
    bad_behaviour_events = events[events['bad_behaviour_card'].notna()]
else:
    bad_behaviour_events = pd.DataFrame()  # Crear un DataFrame vacío si la columna no existe

# 7. Crear un DataFrame vacío para almacenar todas las alineaciones y estadísticas
all_data_df = pd.DataFrame()

# 8. Procesar alineaciones y agregar estadísticas individuales
for team_name, team_lineup in lineups.items():
    team_lineup['team'] = team_name  # Añadir el nombre del equipo
    team_lineup_filtered = team_lineup[['player_id', 'country', 'player_name', 'jersey_number', 'positions', 'team']].copy()
    
    # Extraer la posición inicial de cada jugador
    team_lineup_filtered['initial_position'] = team_lineup_filtered['positions'].apply(
        lambda positions: positions[0]['position'] if isinstance(positions, list) and len(positions) > 0 else None
    )
    team_lineup_filtered.drop(columns=['positions'], inplace=True)  # Eliminar la columna original 'positions'
    
    # Crear columna 'foul_committed_card' solo si hay eventos de tarjetas
    if not bad_behaviour_events.empty:
        team_lineup_filtered['foul_committed_card'] = team_lineup_filtered['player_id'].apply(
            lambda x: ', '.join(
                [f"{card}- {pd.to_datetime(time, unit='s').strftime('%M:%S')}" 
                 for card, time in zip(bad_behaviour_events[bad_behaviour_events['player'] == x]['bad_behaviour_card'], 
                                       bad_behaviour_events[bad_behaviour_events['player'] == x]['timestamp'])]
            )
        )
    else:
        team_lineup_filtered['foul_committed_card'] = ""  # Dejar en blanco si no hay tarjetas

    # 9. Filtrar eventos para el equipo específico
    team_events = events[events['team'] == team_name]

    # 9. Contar eventos específicos por jugador
    ball_recovery_count = team_events[team_events['type'] == 'Ball Recovery'].groupby('player').size().reset_index(name='ball_recovery_count')
    shot_count = team_events[team_events['type'] == 'Shot'].groupby('player').size().reset_index(name='shot_count')
    foul_committed_count = team_events[team_events['type'] == 'Foul Committed'].groupby('player').size().reset_index(name='foul_committed_count')

    # Contar "VERDADERO" en "pass_cross" por jugador
    pass_cross_count = team_events[team_events['pass_cross'] == True].groupby('player').size().reset_index(name='pass_cross_true_count')

    # 10. Crear el DataFrame 'individual_stats' combinando los conteos
    individual_stats = pd.merge(ball_recovery_count, shot_count, on='player', how='outer')
    individual_stats = pd.merge(individual_stats, foul_committed_count, on='player', how='outer')
    individual_stats = pd.merge(individual_stats, pass_cross_count, on='player', how='outer')

    # Rellenar valores nulos con ceros
    individual_stats.fillna(0, inplace=True)

    # Unir alineaciones y estadísticas individuales, eliminando 'player' y dejando 'player_name'
    combined_df = pd.merge(team_lineup_filtered, individual_stats, left_on='player_name', right_on='player', how='left')
    combined_df.drop(columns=['player'], inplace=True)  # Eliminar la columna 'player'
    combined_df.fillna(0, inplace=True)  # Rellenar valores nulos con ceros

    # Combinar "country" y "team" si son la misma información
    if 'country' in combined_df.columns and 'team' in combined_df.columns:
        combined_df.drop(columns=['country'], inplace=True)

    # 11. Ordenar jugadores: primero los que tienen posición, luego los que no
    combined_df['initial_position'] = combined_df['initial_position'].astype(str).replace('nan', '')
    combined_df.sort_values(by=['initial_position', 'player_name'], inplace=True, key=lambda col: col.str.lower())

    # Añadir los datos procesados al DataFrame general
    all_data_df = pd.concat([all_data_df, combined_df], ignore_index=True)

# 12. Crear un nuevo PDF en modo horizontal (landscape)
pdf = FPDF(orientation='L', unit='mm', format='A4')
pdf.set_auto_page_break(auto=True, margin=15)

# 13. Generar una Hoja en el PDF para Cada Equipo
# Procesar cada equipo y generar una hoja en el PDF
for team_name in all_data_df['team'].unique():
    team_data = all_data_df[all_data_df['team'] == team_name]

    # Añadir el nombre del equipo como encabezado
    pdf.add_page()
    pdf.set_font("Arial", 'B', size=14)
    pdf.cell(0, 10, team_name, ln=True, align='C')
    pdf.ln(5)  # Espacio después del encabezado

    # Añadir el encabezado del partido
    pdf.set_font("Arial", size=8)
    header_text = (f"Match Date: {match_date} | Match ID: {match_id} | Home Team: {home_team} "
                   f"({home_score}) vs Away Team: {away_team} ({away_score}) | Stage: {competition_stage}")
    pdf.multi_cell(0, 10, header_text, border=0, align='C')
    pdf.ln(5)  # Espacio después del encabezado del partido

    # Añadir las columnas del DataFrame al PDF
    pdf.set_font("Arial", size=8)
    column_width = max(25, 200 / len(team_data.columns))  # Ajuste dinámico del ancho de columna

    for column in team_data.columns:
        pdf.cell(column_width, 10, str(column), border=1)

    pdf.ln()

    # Añadir las filas del DataFrame al PDF
    for index, row in team_data.iterrows():
        for item in row:
            pdf.cell(column_width, 10, str(item), border=1)
        pdf.ln()

# 14. Guardar el archivo PDF
nombre_archivo_pdf = f"{match_id}_individual_stats.pdf"
ruta_archivo_pdf = os.path.join(ruta_base_pdf, nombre_archivo_pdf)
pdf.output(ruta_archivo_pdf)

print(f"\nArchivo PDF guardado en: {ruta_archivo_pdf}")
```
